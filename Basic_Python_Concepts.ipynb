```python
# data types
print(type(5)) # integer
print(type(5.0)) # float
print(type('five')) # string
print(type(True)) # bool i.e 1 is True, 0 is False
```

```python
# comment
print('sample text') # here is a comment to define or explain code and is ignored
```

```python
# numerial representations
print(0b10) # binary, has b in it
print(0o10) # octal, has o in it
print(0x10) # hexadecimal, has x in it
```

```python
# operators
print(10 + 5) # addition
print(10 - 5) # subtraction
print(10 * 5) # multiplication
print(10 / 5) # division
print(10 % 5) # modulus
print(10 ** 5) # exponent
print(10 // 5) # floor
```

```python
# reassigning values
x = 5
print(x)

x = 10 # reassigning x
print(x)
```

```python
# type casting

# Implicit Type Conversion (Python handles it automatically)
num_int = 10
num_float = 20.5
sum_result = num_int + num_float
print(f"Sum of int and float (implicit): {sum_result}, Type: {type(sum_result)}")

# Explicit Type Conversion (you tell Python to convert)
# int() - converts to integer
string_num = "100"
int_from_string = int(string_num)
print(f"String to int: {int_from_string}, Type: {type(int_from_string)}")

# float() - converts to float
int_to_float = float(num_int)
print(f"Int to float: {int_to_float}, Type: {type(int_to_float)}")

# str() - converts to string
float_to_string = str(num_float)
print(f"Float to string: {float_to_string}, Type: {type(float_to_string)}")

# bool() - converts to boolean (0, None, empty sequences are False, everything else is True)
print(f"int(0) to bool: {bool(0)}, Type: {type(bool(0))}")
print(f"int(5) to bool: {bool(5)}, Type: {type(bool(5))}")
print(f"empty string to bool: {bool('')}, Type: {type(bool(''))}")
print(f"non-empty string to bool: {bool('hello')}, Type: {type(bool('hello'))}")
```
