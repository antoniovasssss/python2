```python
# double quotes
print('hello world!')
```

```python
# single quotes
print('hello world!')
```

```python
# quotes with escape
print('I\'m learning python!')
```

```python
# one line combination with (,)
print('hello world!', 'I love learning python!')
```

```python
# insert new line (\n)
print('I love learning\npython!')
```

```python
# variable
greeting = 'hello, friends!'
print(greeting)
```

```python
# data types
print(type(5)) # integer
print(type(5.0)) # float
print(type('five')) # string
print(type(True)) # bool i.e 1 is True, 0 is False
```

```python
# comment
print('sample text') # here is a comment to define or explain code and is ignored
```

```python
# numerial representations
print(0b10) # binary, has b in it
print(0o10) # octal, has o in it
print(0x10) # hexadecimal, has x in it
```

```python
# operators
print(10 + 5) # addition
print(10 - 5) # subtraction
print(10 * 5) # multiplication
print(10 / 5) # division
print(10 % 5) # modulus
print(10 ** 5) # exponent
print(10 // 5) # floor
```

```python
# reassigning values
x = 5
print(x)

x = 10 # reassigning x
print(x)
```

```python
# the input() function
print('what is your name?')
user_name = input()
print('Hello,', user_name)
```
